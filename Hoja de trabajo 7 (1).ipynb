## Problema 1

Un proyectil que viaja con velocidad $\mathbf{v}_0$ explota en tres pedazos de masas iguales.  
Justo después de la explosión:

- Un pedazo tiene velocidad $\mathbf{v}_1 = \mathbf{v}_0$.  
- Los otros dos pedazos tienen velocidades $\mathbf{v}_2$ y $\mathbf{v}_3$ que cumplen  
  $|\mathbf{v}_2| = |\mathbf{v}_3|$ y $\mathbf{v}_2 \perp \mathbf{v}_3$.

Determinar $\mathbf{v}_2$ y $\mathbf{v}_3$ en función de $\mathbf{v}_0$.  
Esbozar las tres velocidades en el plano (vectorialmente).

Notas: la masa inicial es $M$ y las tres masas resultantes son $m = M/3$.  
Se conserva el momento lineal total.



### Solución
Usando conservación del momento lineal:


$M \mathbf{v}_0 = 3m \mathbf{v}_{0} = m (\mathbf{v}_1 + \mathbf{v}_2 + \mathbf{v}_3)$


Como $m = M/3$:

$3\mathbf{v}_0 = \mathbf{v}_1 + \mathbf{v}_2 + \mathbf{v}_3$


Sustituyendo $\mathbf{v}_1 = \mathbf{v}_0$:

$2\mathbf{v}_0 = \mathbf{v}_2 + \mathbf{v}_3$


Si $\mathbf{v}_2$ y $\mathbf{v}_3$ son perpendiculares e iguales en magnitud, y su suma es paralela a $\mathbf{v}_0$,  
entonces cada una forma un ángulo de 45° con la dirección de $\mathbf{v}_0$.

$|\mathbf{v}_2| = |\mathbf{v}_3| = \frac{2v_0}{\sqrt{2}} = \sqrt{2}\,v_0$



```python
import matplotlib.pyplot as plt
import numpy as np

v0 = 1
v1 = np.array([v0, 0])
v2 = np.array([v0/np.sqrt(2), v0/np.sqrt(2)])
v3 = np.array([v0/np.sqrt(2), -v0/np.sqrt(2)])

plt.figure(figsize=(6,6))
plt.quiver(0,0,*v1,angles='xy',scale_units='xy',scale=1,color='r',label=r'$\mathbf{v}_1$')
plt.quiver(0,0,*v2,angles='xy',scale_units='xy',scale=1,color='g',label=r'$\mathbf{v}_2$')
plt.quiver(0,0,*v3,angles='xy',scale_units='xy',scale=1,color='b',label=r'$\mathbf{v}_3$')
plt.xlim(-0.2,1.5)
plt.ylim(-1,1)
plt.grid(True)
plt.legend()
plt.title("Velocidades después de la explosión")
plt.show()
```

## Problema 2

Un asteroide esférico de radio $R_0$ está girando con velocidad angular $\omega_0$.  
Con el paso del tiempo recoge materia hasta que su radio es $R$,  
manteniendo densidad constante.  
La materia adicional estaba en reposo relativo al asteroide.

Encontrar la nueva velocidad angular $\omega$ y,  
si $R$ se dobla, determinar $\omega_f$.

### Solución
El momento de inercia de una esfera sólida es:

$I = \frac{2}{5} M R^2$

Con densidad constante, $M \propto R^3$, por lo tanto:

$I \propto R^5$

Por conservación del momento angular:

$I_0 \omega_0 = I \omega$

Entonces:

$\omega = \omega_0 \left(\frac{R_0}{R}\right)^5$

Si el radio se duplica ($R = 2R_0$):

$\omega_f = \frac{\omega_0}{32}$


## Problema 3

Una granada se lanza desde el origen con velocidad inicial  
$\mathbf{v}_0 = (4, 4)$ y $g = 1$.

1. Representar su movimiento para $0 \le t \le 4$ y marcar las posiciones en $t=1,2,3,4$.  
2. En $t=4$, cuando la velocidad es $\mathbf{v}$, la granada explota en dos pedazos iguales.  
   Uno adquiere velocidad $\mathbf{v} + \Delta\mathbf{v}$ y el otro debe mantener el centro de masa.  
3. Si $\Delta\mathbf{v} = (1, 3)$, graficar las trayectorias de ambos pedazos para $4 < t \le 9$,  
   mostrando sus posiciones en $t = 5, 6, 7, 8, 9$ y comprobando que el CM sigue la trayectoria parabólica original.


### Solución 

Ecuaciones del movimiento:

$x(t) = v_{0x} t, \quad y(t) = v_{0y} t - \frac{1}{2} g t^2$

En $t=4$:


$\mathbf{v}(4) = (v_{0x}, v_{0y} - g t)$


Después de la explosión, los fragmentos tienen velocidades:

$\mathbf{v}_1 = \mathbf{v} + \Delta\mathbf{v}, \quad
\mathbf{v}_2 = \mathbf{v} - \Delta\mathbf{v}$


```python
import numpy as np
import matplotlib.pyplot as plt

g = 1.0
v0 = np.array([4.0, 4.0])
r0 = np.array([0.0, 0.0])
a = np.array([0.0, -g])


t_full = np.linspace(0, 4, 201)
positions = np.array([r0 + v0*t + 0.5*a*t*t for t in t_full])
t_marks = np.array([1,2,3,4], dtype=float)
pos_marks = np.array([r0 + v0*t + 0.5*a*t*t for t in t_marks])


t_exp = 4.0
v_t4 = v0 + a*t_exp
r_t4 = r0 + v0*t_exp + 0.5*a*t_exp*t_exp

Delta_v = np.array([1.0, 3.0])
v_piece1 = v_t4 + Delta_v
v_piece2 = v_t4 - Delta_v

t_post = np.linspace(4, 9, 501)
pos_piece1 = np.array([r_t4 + v_piece1*(t-4) + 0.5*a*(t-4)**2 for t in t_post])
pos_piece2 = np.array([r_t4 + v_piece2*(t-4) + 0.5*a*(t-4)**2 for t in t_post])
pos_cm_post = 0.5*(pos_piece1 + pos_piece2)

t_full_long = np.linspace(0, 9, 401)
pos_original_long = np.array([r0 + v0*t + 0.5*a*t*t for t in t_full_long])

t_marks_post = np.array([5,6,7,8,9], dtype=float)
pos_piece1_marks = np.array([r_t4 + v_piece1*(t-4) + 0.5*a*(t-4)**2 for t in t_marks_post])
pos_piece2_marks = np.array([r_t4 + v_piece2*(t-4) + 0.5*a*(t-4)**2 for t in t_marks_post])
pos_cm_marks = 0.5*(pos_piece1_marks + pos_piece2_marks)


plt.figure(figsize=(8,6))
plt.plot(positions[:,0], positions[:,1], label='Trayectoria (0≤t≤4)')
plt.scatter(pos_marks[:,0], pos_marks[:,1], marker='x', label='Posiciones t=1,2,3,4')
for i,t in enumerate(t_marks):
    plt.annotate(f"t={int(t)}", (pos_marks[i,0], pos_marks[i,1]), textcoords="offset points", xytext=(5,5))
plt.axhline(0, linewidth=0.5)
plt.axvline(0, linewidth=0.5)
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Trayectoria del proyectil antes de explosión (g=1, v0=(4,4))')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()


plt.figure(figsize=(8,6))
plt.plot(pos_piece1[:,0], pos_piece1[:,1], label='Pieza 1 (t≥4)')
plt.plot(pos_piece2[:,0], pos_piece2[:,1], label='Pieza 2 (t≥4)')
plt.plot(pos_original_long[:,0], pos_original_long[:,1], '--', label='Trayector. original (0≤t≤9)')
plt.plot(pos_cm_post[:,0], pos_cm_post[:,1], ':', linewidth=2, label='CM (t≥4)')
plt.scatter(pos_piece1_marks[:,0], pos_piece1_marks[:,1], marker='o')
plt.scatter(pos_piece2_marks[:,0], pos_piece2_marks[:,1], marker='s')
plt.scatter(pos_cm_marks[:,0], pos_cm_marks[:,1], marker='+')
for i,t in enumerate(t_marks_post):
    plt.annotate(f"t={int(t)}", (pos_piece1_marks[i,0], pos_piece1_marks[i,1]), textcoords="offset points", xytext=(5,5))
    plt.annotate(f"t={int(t)}", (pos_piece2_marks[i,0], pos_piece2_marks[i,1]), textcoords="offset points", xytext=(5,-12))
    plt.annotate(f"t={int(t)}", (pos_cm_marks[i,0], pos_cm_marks[i,1]), textcoords="offset points", xytext=(-20,5))
plt.scatter([r_t4[0]],[r_t4[1]], marker='*', s=100, label='Explosión en t=4')
plt.gca().set_aspect('equal', adjustable='box')
plt.title('Trayectorias tras explosión (Δv=(1,3)) y CM vs trayectoria original')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid(True)
plt.show()
```
